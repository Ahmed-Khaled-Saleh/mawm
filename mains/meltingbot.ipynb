```python
from shimmy import MeltingPotCompatibilityV0
from shimmy.utils.meltingpot import load_meltingpot
import numpy as np

# 1. Load environment
env = load_meltingpot("collaborative_cooking__circuit")
env = MeltingPotCompatibilityV0(env, render_mode=None)

# 2. Reset environment
reset_out = env.reset()
if isinstance(reset_out, tuple):  # unpack tuple (obs, info)
    observations, infos = reset_out
else:
    observations = reset_out
    infos = None

# 3. Initialize data storage
data = {agent: {"obs": [], "actions": [], "rewards": []} for agent in env.agents}

# 4. Collect random data
while env.agents:
    actions = {}
    for agent in env.agents:
        action = env.action_space(agent).sample()
        actions[agent] = action

        # access observation dict correctly
        obs = observations[agent]
        data[agent]["obs"].append(obs)
        data[agent]["actions"].append(action)

    # perform step
    step_out = env.step(actions)
    if len(step_out) == 5:
        observations, rewards, terminations, truncations, infos = step_out
    else:
        observations, rewards, terminations, truncations = step_out
        infos = None

    # store rewards
    for agent, r in rewards.items():
        data[agent]["rewards"].append(r)

env.close()

# 5. Convert to numpy arrays
for agent, d in data.items():
    d["obs"] = np.array(d["obs"])
    d["actions"] = np.array(d["actions"])
    d["rewards"] = np.array(d["rewards"])

# 6. Print summary
for agent, d in data.items():
    print(f"{agent}: {len(d['obs'])} steps, avg reward = {np.mean(d['rewards']):.3f}")
```

```python
from shimmy import MeltingPotCompatibilityV0
from shimmy.utils.meltingpot import load_meltingpot
import numpy as np

# 1. Load environment
env = load_meltingpot("collaborative_cooking__circuit")
env = MeltingPotCompatibilityV0(env, render_mode=None)

# 2. Reset
reset_out = env.reset()
if isinstance(reset_out, tuple):
    observations, infos = reset_out
else:
    observations = reset_out
    infos = None

# 3. Access the underlying DeepMind Melting Pot env
dm_env = env.unwrapped._env  # internal raw env
    
# 4. Initialize data storage
data = {
    "global_state": [],
    "agents": {
        agent: {"obs": [], "actions": [], "rewards": []}
        for agent in env.agents
    },
}

# 5. Rollout loop
while env.agents:
    actions = {}
    for agent in env.agents:
        action = env.action_space(agent).sample()
        actions[agent] = action
        obs = observations[agent]
        data["agents"][agent]["obs"].append(obs)
        data["agents"][agent]["actions"].append(action)

    # get global state *before* the step
    global_state = env.state()  # numpy array (H, W, Channels)
    # print(global_state.shape)
    data["global_state"].append(global_state)

    step_out = env.step(actions)
    if len(step_out) == 5:
        observations, rewards, terminations, truncations, infos = step_out
    else:
        observations, rewards, terminations, truncations = step_out
        infos = None

    # store rewards
    for agent, r in rewards.items():
        data["agents"][agent]["rewards"].append(r)

env.close()

# 6. Convert to numpy arrays
for agent, d in data["agents"].items():
    d["obs"] = np.array(d["obs"])
    d["actions"] = np.array(d["actions"])
    d["rewards"] = np.array(d["rewards"])

data["global_state"] = np.array(data["global_state"])

# 7. Print summary
for agent, d in data["agents"].items():
    print(f"{agent}: {len(d['obs'])} steps, avg reward = {np.mean(d['rewards']):.3f}")

print("Global states recorded:", data["global_state"].shape)
```
